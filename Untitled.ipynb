```python
width = 5.0
Wx = 1.0
Wy = 1.0
b = -1.2
alpha = 0.4
plt.rcParams['figure.figsize'] = [6, 12]
fig = plt.figure()
ax = fig.add_subplot(211, label='2d')
ax2 = fig.add_subplot(212, projection='3d', label='3d')

#fig, ax = plt.subplots()

#plt.subplots_adjust(left=0.25)
#rax = plt.axes([0, 0.2, 0.2, 0.7])
#plt.rc('text', usetex=True)

lines = []
labels=[]
visibility=[]
img = None

def func(label):
    index = labels.index(label)
    lines[index].set_visible(not lines[index].get_visible())
    plt.draw()

def modifier(event):
    global Wx
    global Wy
    global b
    global alpha
    global width
    global lines
    global labels
    global visibility
    global img
    
    color = iter(['r', 'g', 'b', 'y'])
    
    if event:
        if event.key == '.':
            Wx = Wx + 0.1
            if Wx == 0:
                Wx = 0.1
        elif event.key == ',':
            Wx = Wx - 0.1
            if Wx == 0:
                Wx = -0.1
        elif event.key == 'l':
            Wy = Wy + 0.1
            if Wy == 0:
                Wy = 0.1
        elif event.key == 'k':
            Wy = Wy - 0.1
            if Wy == 0:
                Wy = -0.1
        elif event.key == "o":
            b = b + 0.1
        elif event.key == 'i':
            b = b - 0.1
        elif event.key == "j":
            alpha = alpha + 0.1
        elif event.key == 'h':
            alpha = alpha - 0.1
        
    WTrue = np.array([Wx, Wy])
    xBig = np.array([-width, width])

    W = np.copy(WTrue)
    ax.clear()
    for i in itertools.product([True, False], repeat=2):
        xBig = np.array([-width, width])
        c = next(color)
        i = list(i)
        if i[0]:
            x = np.linspace(-width, width, 1000)
        else:
            x = np.linspace(-width, width, 1000)
        W[i] = alpha*W[i]
        
        if W[1] != 0:
            y = (-b - W[0]*x) / W[1]
            yBig = (-b - W[0]*xBig) / W[1]
        elif W[0] == 0:
            x = np.array([0])
            xBig = np.array([0])
            y = np.array([0])
            yBig = np.array([0])
        else:
            xBig = -b/W[0]*np.ones(1000)
            yBig = np.linspace(-width, width, 1000)
            x = -b/W[0]*np.ones(1000)
            y = np.linspace(-width, width, 1000)
            
        
        if i[0] and i[1]:
            idx = np.logical_and(x <= width/100, y <= width/100)
            eq = -b/(W[0] + W[1]) if W[0] != -W[1] else 0
            if np.linalg.norm(W) > 0:
                direct = W/np.linalg.norm(W)
                ax.arrow(eq, eq, direct[0], direct[1], color=c, head_width=0.05)
            #plt.fill_between(x, y, -width, where=idx, color=c, alpha=0.1)
            x = x[idx]
            y = y[idx]
            label = 'x, y < 0'
        elif not i[0] and i[1]:
            idx = np.logical_and(x >= -width/1000, y <= width/1000)
            direct = W/np.linalg.norm(W)
            if W[1] != 0:
                ax.arrow(-b/W[0]/2, -b/W[1]/2, direct[0], direct[1], color=c, head_width=0.05)
            else:
                ax.arrow(-b/W[0], -width/2, direct[0], direct[1], color=c, head_width=0.05)
            #plt.fill_between(x, y, -width, where=idx, color=c, alpha=0.1)
            x = x[idx]
            y = y[idx]
            label = 'x > 0, y < 0'
        elif i[0] and not i[1]:
            idx = np.logical_and(x <= width/1000, y >= -width/1000)
            direct = W/np.linalg.norm(W)
            if W[0] != 0:
                ax.arrow(-b/W[0]/2, -b/W[1]/2, direct[0], direct[1], color=c, head_width=0.05)
            else:
                ax.arrow(-width/2, -b/W[1], direct[0], direct[1], color=c, head_width=0.05)
            #plt.fill_between(x, y, -width, where=idx, color=c, alpha=0.1)
            x = x[idx]
            y = y[idx]
            label = 'x < 0, y > 0'
        elif not i[0] and not i[1]:
            idx = np.logical_and(x >= -width/1000, y >= -width/1000)
            eq = -b/(W[0] + W[1])
            direct = W/np.linalg.norm(W)
            ax.arrow(eq, eq, direct[0], direct[1], color=c, head_width=0.05)
            #plt.fill_between(x, y, width, where=idx, color=c, alpha=0.1)
            x = x[idx]
            y = y[idx]
            label = 'x, y > 0'
        ax.plot(xBig, yBig, color=c, label=label)
        l, = ax.plot(x, y, color='k')
        W = np.copy(WTrue)
        lines.append(l)
    
    ax.set_title(r'W_1 = {:3.1f}, W_2 = {:3.1f}, b = {:3.1f}, \alpha = {:3.1f}'.format(Wx, Wy, b, alpha))
    ax.set_ylim([-width, width])
    ax.set_xlim([-width, width])
    ax.legend()
    ax.set_xticks([0])
    ax.set_yticks([0])
    ax.grid(True)
    ax.set_xlabel('s_1')
    ax.set_ylabel('s_2')
    
    x = np.linspace(-width, width, 100)
    y = np.copy(x)

    X, Y = np.meshgrid(x, y)
    origShape = X.shape

    X = np.reshape(X, -1)
    Y = np.reshape(Y, -1)

    bothIdx = np.logical_and(X >= 0, Y >= 0)
    neitherIdx = np.logical_and(X <= 0, Y <= 0)
    xIdx = np.logical_and(X <= 0, Y >= 0)
    yIdx = np.logical_and(X >= 0, Y <= 0)

    Z = np.zeros(X.shape)

    Z[bothIdx] = Wx*X[bothIdx] + Wy*Y[bothIdx] + b
    Z[neitherIdx] = alpha*(Wx*X[neitherIdx] + Wy*Y[neitherIdx]) + b
    Z[xIdx] = alpha*Wx*X[xIdx] + Wy*Y[xIdx] + b
    Z[yIdx] = Wx*X[yIdx] + alpha*Wy*Y[yIdx] + b

    bounds = np.linspace(-1, 1, 3)
    norm = matplotlib.colors.BoundaryNorm(boundaries=bounds, ncolors=256)

    X = np.reshape(X, origShape)
    Y = np.reshape(Y, origShape)
    Z = np.reshape(Z, origShape)

    ax2.clear()
    
    img = ax2.plot_surface(X, Y, Z, cmap='RdBu_r', vmin=-1, vmax=1, linewidth=0, antialiased=False);
    #img = ax2.plot_surface(X, Y, Z, cmap='RdBu_r', norm=norm, linewidth=0, antialiased=False);
    
    ax2.scatter([width], [width], [Wx*width + Wy*width + b], color='y')
    ax2.scatter([width], [-width], [Wx*width - alpha*Wy*width + b], color='b')
    ax2.scatter([-width], [width], [-alpha*Wx*width + Wy*width + b], color='g')
    ax2.scatter([-width], [-width], [-alpha*Wx*width - alpha*Wy*width + b], color='r')
    
    ax2.plot([0, width], [0, 0], [b, Wx*width + b], color='k')
    ax2.plot([0, 0], [0, width], [b, Wy*width + b], color='k')
    ax2.plot([0, -width], [0, 0], [b, -alpha*Wx*width + b], color='k')
    ax2.plot([0, 0], [0, -width], [b, -alpha*Wy*width + b], color='k')

    ax2.set_title(r'W_1 = {:3.1f}, W_2 = {:3.1f}, b = {:3.1f}, \alpha = {:3.1f}'.format(Wx, Wy, b, alpha))
    ax2.set_ylim([-width, width])
    ax2.set_xlim([-width, width])
    ax2.set_zlim([- np.amax(abs(Z)), np.amax(abs(Z))])
    ax2.set_xticks([-width, 0, width])
    ax2.set_yticks([-width, 0, width])
    ax2.set_zticks([-np.amax(abs(Z)), 0, np.amax(abs(Z))])
    ax2.set_xlabel('s_1')
    ax2.set_ylabel('s_2')
    ax2.set_zlabel('W*ReLU(s) + b')

#     labels = [str(line.get_label()) for line in lines]
#     visibility = [line.get_visible() for line in lines]
#     check = CheckButtons(rax, labels, visibility)

#     check.on_clicked(func)

modifier(None)
    
fig.canvas.mpl_connect('key_press_event', modifier)
fig.colorbar(img, extend='both');

plt.show()
```

```python
width = 5.0

W = np.array([1.0, 1.0])
bo = -0.4
V = np.array([[1.0, 0.0], [0.0, 1.0]])
bi = np.array([[1.0], [1.0]])

alpha = 0.4
plt.rcParams['figure.figsize'] = [10, 10]
fig = plt.figure()
ax = fig.add_subplot(111)

def ReLU(x, alpha):
    ret = np.copy(x)
    ret[ret < 0] = alpha*ret[ret < 0]
    return ret

x = np.linspace(-width, width, 1000)
y = np.linspace(-width, width, 1000)

X, Y = np.meshgrid(x, y)
X = np.reshape(X, -1)
Y = np.reshape(Y, -1)

pts = np.c_[X, Y].T

O = np.dot(W, ReLU(np.dot(V, ReLU(pts, alpha)) + bi, alpha)) + bo

pos = O > 0
neg = O < 0

plt.plot(X[pos], Y[pos], 'ro', markersize=1)
plt.plot(X[neg], Y[neg], 'bo', markersize=1)
```

```python
width = 10.0

Wx = 1.0
Wy = 1.0
bo = -0.4

Vx1 = 1.0
Vy1 = 0.4
Vx2 = 0.4
Vy2 = 1.0
bix = 0.0
biy = 0.0

alpha = 0.4

plt.rcParams['figure.figsize'] = [10, 10]
fig = plt.figure()
ax = fig.add_subplot(111)

x = np.linspace(-width, width, 5000)

def checkIn(xSign, ySign, gamma3, gamma4, x, y):
    global Vx1
    global Vx2
    global Vy1
    global Vy2
    global bix
    global biy
    
    if xSign:
        idx = x >= 0
    else:
        idx = x <= 0
        
    if ySign:
        idx = np.logical_and(idx, y >= 0)
    else:
        idx = np.logical_and(idx, y <= 0)

    return idx
        
def checkInFull(xSign, ySign, sxSign, sySign, gamma3, gamma4, x, y):
    global Vx1
    global Vx2
    global Vy1
    global Vy2
    global bix
    global biy
    
    idx = checkIn(xSign, ySign, gamma3, gamma4, x, y)
    
    sx = gamma3*Vx1*x + gamma4*Vy1*y + bix
    sy = gamma3*Vx2*x + gamma4*Vy2*y + biy
    
    if sxSign:
        idx = np.logical_and(idx, sx >= 0)
    else:
        idx = np.logical_and(idx, sx <= 0)
        
    if sySign:
        idx = np.logical_and(idx, sy >= 0)
    else:
        idx = np.logical_and(idx, sy <= 0)

    return idx

for i in itertools.product([True, False], repeat=2):
        i = list(i)
        
        if i[0]:
            gamma3 = 1
            label = 'x > 0, '
        else:
            gamma3 = alpha
            label = 'x < 0, '
            
        if i[1]:
            gamma4 = 1
            label += 'y > 0, '
        else:
            gamma4 = alpha
            label += 'y < 0, '
            
        if Vy1 != 0:
            sxY = (-bix - gamma3*Vx1*x)/(gamma4*Vy1)
            idx = checkIn(i[0], i[1], gamma3, gamma4, x, sxY)
            ax.plot(x[idx], sxY[idx], 'k')
        else:
            sxX = (-bix - gamma4*Vy1*x)/(gamma3*Vx1)
            idx = checkIn(i[0], i[1], gamma3, gamma4, sxX, x)
            ax.plot(sxX[idx], x[idx], 'k')
            
        if Vy2 != 0:
            syY = (-biy - gamma3*Vx2*x)/(gamma4*Vy2)
            idx = checkIn(i[0], i[1], gamma3, gamma4, x, syY)
            ax.plot(x[idx], syY[idx], 'k')
        else:
            syX = (-bix - gamma4*Vy1*x)/(gamma3*Vx1)
            idx = checkIn(i[0], i[1], gamma3, gamma4, syX, x)
            ax.plot(syX[idx], x[idx], 'k')
        
        
        for j in itertools.product([True, False], repeat=2):
            j = list(j)
            
            if j[0]:
                gamma1 = 1
                label2 = label + 's_x > 0, '
            else:
                gamma1 = alpha
                label2 = label + 's_x < 0, '
                
            if j[1]:
                gamma2 = 1
                label2 += 's_y > 0'
            else:
                gamma2 = alpha
                label2 += 's_y < 0'
                
            y = (-Wx*gamma1*bix - Wy*gamma2*biy - bo - gamma3*(Wx*gamma1*Vx1 + Wy*gamma2*Vx2)*x)/(gamma4*(Wx*gamma1*Vy1 + Wy*gamma2*Vy2))
            
            idx = checkInFull(i[0], i[1], j[0], j[1], gamma3, gamma4, x, y)
            
            if np.any(idx):
                ax.plot(x[idx], y[idx], label=label2)
                
            
     
    
W = np.array([Wx, Wy])
V = np.array([[Vx1, Vy1], [Vx2, Vy2]])

ax.set_title(r'$W = {} {}$'.format(Wx, Wy), loc='left', family='monospace')
ax.set_ylim([-width, width]);
ax.set_xlim([-width, width]);
ax.legend();
ax.set_xticks([0]);
ax.set_yticks([0]);
ax.grid(True);
ax.set_xlabel('x');
ax.set_ylabel('y');
```

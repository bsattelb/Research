```python
temp = np.load('saveFile.npz')
A1 = temp['A1']
A2 = temp['A2']
b1 = temp['b1']
b2 = temp['b2']
```

```python
Fterms, Gterms = tropical.getTropCoeffs(A1, b1, A2, b2)
```

```python
display(Math('F^{(2)} = ' + tropical.displayTropPoly(Fterms)))
print()
display(Math('G^{(2)} = ' + tropical.displayTropPoly(Gterms)))
```

```python
fig = plt.figure(figsize=(5, 10))
ax = fig.add_subplot(211, projection='3d')
tropical.newtonPolygon(Fterms, ax)
ax.set_title('Newton polygon of F');

ax = fig.add_subplot(212, projection='3d')
tropical.newtonPolygon(Gterms, ax)
ax.set_title('Newton polygon of G');
```
